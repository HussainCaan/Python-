### Joining Arrays
#### VERTICAL STACK |  HORIZONTAL STACK | COLUMN STACK

```python
newarray1 = np.array([1, 2, 3])
newarray2 = np.array([1, 2, 3])
joined_array = np.vstack((newarray1, newarray2))
joined_array
```

```python
newarray3 = np.array([[1, 2, 3], [4, 5, 6]])
newarray4 = np.array([[7, 8, 9], [10, 11, 12]])
joined_array2 = np.hstack((newarray3, newarray4))
joined_array2
```

```python
newarray5 = np.array([[1, 2, 3], [4, 5, 6]])
newarray6 = np.array([[7, 8, 9], [10, 11, 12]])
joined_array3 = np.column_stack((newarray5, newarray6))
joined_array3
```

### Time:  NUMPY ARRAY ARE FASTER

```python
import time
```

```python
SIZE = 1000000

list1 = list(range(SIZE))
list2 = list(range(SIZE))

array1 = np.arange(SIZE)
array2 = np.arange(SIZE)

```

```python
List_start_time = time.time()
list3 = list1 + list2
List_end_time = time.time()
print("Time taken for list addition:", List_end_time - List_start_time)
Array_start_time = time.time()
array3 = array1 + array2
Array_end_time = time.time()
print("Time taken for array addition:", Array_end_time - Array_start_time)
print("Are the results equal?", np.array_equal(list3, array3))
```

# SQUARE ROOT AND STANDARD DEVIATION

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
print("Original array:\n", a)
print("SQUARE ROOT of the array:\n", np.sqrt(a))
print("SQUARE of the array:\n", np.square(a))
print("STANDARD DEVIATION of the array:\n", np.std(a))
print("MEAN of the array:\n", np.mean(a))
```
