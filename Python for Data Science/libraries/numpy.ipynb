```python
import numpy as np
```

```python
array1 = np.array([1, 2, 3, 4, 5])
print("Array:", array1)
print("Shape:", array1.shape)
print("Data Type:", array1.dtype)
```

```python
array2 = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", array2)
print("Shape:", array2.shape)
print("Data Type:", array2.dtype)
```
