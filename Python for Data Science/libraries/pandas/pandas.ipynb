```python
# Three dictionaries for merge practice
dict_a = {"id": 1, "name": "Alice", "skills": ["Python", "SQL"]}
dict_b = {"id": 1, "email": "alice@example.com", "location": "NY"}
dict_c = {"id": 1, "name": "Alice B.", "skills": ["Python", "Pandas"], "phone": "555-1234"}
```

```python
df1 = pd.DataFrame([dict_a])
df2 = pd.DataFrame([dict_b])
df3 = pd.DataFrame([dict_c])
```

```python
merge = pd.merge(df1, df2, on="id")
```

```python
merge
```

```python
final_df = pd.merge(merge, df3, on="id")
final_df
```
